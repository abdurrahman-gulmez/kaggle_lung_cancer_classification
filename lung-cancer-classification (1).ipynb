```python
from sklearn.preprocessing import OrdinalEncoder

categorical_columns = ["psychologic", "respiratory_disease"]

categories = [['low', 'medium', 'high']]

ordinal_encoder = OrdinalEncoder(categories=categories * len(categorical_columns)) 
cancer_data[categorical_columns] = ordinal_encoder.fit_transform(cancer_data[categorical_columns])
```

```python
from sklearn.model_selection import train_test_split
X = cancer_data.drop(labels=["lung_cancer"],axis=1)
Y = cancer_data["lung_cancer"]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42,stratify=Y)
```

```python
random_state = 123
from sklearn.metrics import accuracy_score
from lightgbm import LGBMClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import BernoulliNB
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
```

HYPERPARAMETER OPTIMIZATION

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import StratifiedKFold
import numpy as np

# StratifiedKFold cross-validation ayarları
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state)

# Modeller ve hiperparametre aralıkları
param_grids = {
    'CatBoostClassifier': {
        'iterations': [100, 200, 500],
        'depth': [4, 6, 8],
        'learning_rate': [0.01, 0.05, 0.1],
        'l2_leaf_reg': [1, 3, 5],
    },
    'LGBMClassifier': {
        'n_estimators': [100, 200, 500],
        'num_leaves': [31, 50],
        'learning_rate': [0.01, 0.05, 0.1],
        'boosting_type': ['gbdt', 'dart'],
        'max_depth': [-1, 3, 5, 7],
        'min_child_samples': [20, 50, 100]
    },
    'LogisticRegression': {
        'penalty': ['l1', 'l2', 'elasticnet'],
        'C': np.logspace(-4, 4, 20),
        'solver': ['saga', 'liblinear'],
        "max_iter": [100, 200, 300]
    },
    'RandomForestClassifier': {
        "criterion" : ["gini","entropy"],
        "max_features": [1,3,10],
        'n_estimators': [100, 200, 500],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        "bootstrap":[False],
        "oob_score": [False]
    },
    'GradientBoostingClassifier': {
        'n_estimators': [100, 200, 500],
        'learning_rate': [0.01, 0.05, 0.1],
        'subsample': [0.5, 0.7, 1.0],
        'max_depth': [3, 5, 7],
    },
    'AdaBoostClassifier': {
        'n_estimators': [50, 100, 200],
        'learning_rate': [0.01, 0.1, 1],
    },
    'DecisionTreeClassifier': {
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        "criterion": ["gini","entropy"]
    },
    'KNeighborsClassifier': {
        'n_neighbors': [3, 5, 7, 9],
        'weights': ['uniform', 'distance'],
        'p': [1, 2],
        "metric":["euclidean","manhattan","minkowski"]
    },
    'SVC': {
        'C': [0.1, 1, 10, 100],
        'kernel': ['linear', 'rbf','poly'],
        'gamma': ['scale', 'auto'],
    },
    'MLPClassifier': {
        'hidden_layer_sizes': [(50,), (100,), (150,)],
        'activation': ['tanh', 'relu','sigmoid','logistic'],
        'solver': ['adam', 'sgd'],
        'alpha': [0.0001, 0.001, 0.01],
        'learning_rate': ['constant', 'invscaling', 'adaptive']
    },
    'BernoulliNB': {
        'alpha': [0.01, 0.1, 1, 10],
        'binarize': [0.0, 0.1, 0.2, 0.5, 1.0],
        'fit_prior': [True, False],
    },
    'XGBClassifier': {
        'n_estimators': [100, 200, 500],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 5, 7],
        'gamma': [0, 1, 5],
        'min_child_weight': [1, 3, 5],
        'subsample': [0.7, 0.8, 0.9],
    }
}

# Modellerin listesi
models = {
    'CatBoostClassifier': CatBoostClassifier(random_state=random_state, verbose=0),
    'LGBMClassifier': LGBMClassifier(random_state=random_state),
    'LogisticRegression': LogisticRegression(random_state=random_state),
    'RandomForestClassifier': RandomForestClassifier(random_state=random_state),
    'GradientBoostingClassifier': GradientBoostingClassifier(random_state=random_state),
    'AdaBoostClassifier': AdaBoostClassifier(random_state=random_state),
    'DecisionTreeClassifier': DecisionTreeClassifier(random_state=random_state),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'SVC': SVC(random_state=random_state),
    'MLPClassifier': MLPClassifier(random_state=random_state),
    'BernoulliNB': BernoulliNB(),
    'XGBClassifier': XGBClassifier(random_state=random_state)
}

# RandomizedSearchCV sonuçlarını saklamak için liste
best_params_list = []

# Her model için RandomizedSearchCV işlemi
for model_name, model in models.items():
    print(f"Performing RandomizedSearchCV for {model_name}")
    param_grid = param_grids[model_name]
    random_search = RandomizedSearchCV(estimator=model, param_distributions=param_grid, 
                                       n_iter=10, scoring='accuracy', cv=cv, verbose=1, n_jobs=-1, random_state=random_state)
    
    # Eğitimi gerçekleştir
    random_search.fit(X_train, y_train)
    
    y_pred = random_search.predict(X_test)
    accuracy = accuracy_score(y_test,y_pred)
    
    # En iyi parametreleri ve skoru sakla
    best_params_list.append({
        'Model': model_name,
        'Best Params': random_search.best_params_,
        'Best Score': random_search.best_score_,
        'Accuracy': accuracy
    })


```

```python
score = 0
acc = 0
best_model = ""
for result in best_params_list:
    if result["Accuracy"] > acc:
        score = result["Best Score"]
        acc = result["Accuracy"]
        best_model = result

print(best_model)
print(f"Best Score: {score}")
print(f"Accuracy: {acc}")
    

```

OPTIMIZATION WITH DEFAULT PARAMETERS

```python
accuracy_values = []

def accuracy_value(model):
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test,y_pred)
    accuracy_values.append({type(model).__name__: accuracy})
    
models = [
    CatBoostClassifier(random_state=random_state, verbose=0),
    LGBMClassifier(random_state=random_state),
    LogisticRegression(random_state=random_state),
    RandomForestClassifier(random_state=random_state),
    GradientBoostingClassifier(random_state=random_state),
    AdaBoostClassifier(random_state=random_state),
    DecisionTreeClassifier(random_state=random_state),
    KNeighborsClassifier(),
    SVC(random_state=random_state),
    MLPClassifier(random_state=random_state),
    BernoulliNB(),
    XGBClassifier(random_state=random_state)
]

for model in models:
    accuracy_value(model)
```

```python

for values in accuracy_values:
    print(values)
```

```python
acc = 0
best_model = ""
for value in accuracy_values:
    model_name, model_acc = list(value.items())[0]
    if model_acc > acc:
        acc = model_acc
        best_model = value

print(best_model)
print(f"Best Accuracy: {acc}")
```
